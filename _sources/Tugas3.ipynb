# Tugas 3 - Crawling data Berita Online

Crawling data berita online (detik.com) menggunakan BeautifulSoup

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
import csv

hades = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'}
```

```python
#membuat function scraping data web
def scrape_berita(page):
  global hades  
  data = []
  for p in range(1,page+1):
    url = f'https://www.detik.com/tag/pemilu-2024/?sortby=time&page={p}'
    request =  req.get(url,hades).text
    soup = bs(request, 'lxml')
    berita_list = soup.find('div', 'list media_rows list-berita')
    berita_article = berita_list.find_all('article')
    for x in berita_article:
      link=x.find_all('h2')
      link = [ele.text.strip() for ele in link]
      content=x.find_all('p')
      content = [ele.text.strip() for ele in content]
      data.append([link[0],content[0]])
  return data

berita=scrape_berita(5)
berita_result = pd.DataFrame(berita,columns =['Judul','Isi'])
berita_result
```

```python
berita_result.to_csv('Berita_Online.csv')
```
