```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
tweet_list = []
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('puan maharani since:2022-01-01 until:2023-02-22').get_items()):
  if i>500:
    break
  tweet_list.append([tweet.user.username,tweet.content,tweet.user.location])
```

```python
frame_tweet = pd.DataFrame(tweet_list,columns =['Akun','Konten','Lokasi'])
frame_tweet
```

```python
frame_tweet.to_csv('tweet_puan.csv')
```
